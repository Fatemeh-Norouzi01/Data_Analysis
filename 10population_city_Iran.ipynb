```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
cities = [
    {"name": "Tehran", "lat": 35.4120, "lon": 51.2323, "population": 9.8},
    {"name": "Mashhad", "lat": 36.1935, "lon": 59.3236, "population": 3.40},
    {"name": "Isfahan", "lat": 32.3955, "lon": 51.4013, "population": 2.2},
    {"name": "Karaj", "lat": 35.4945, "lon": 50.5803, "population": 1.59},
    {"name": "Shiraz", "lat": 29.3636, "lon": 52.3233, "population": 1.9},
    {"name": "Tabriz", "lat": 38.0454, "lon": 46.1757, "population": 1.7},
    {"name": "Qom", "lat": 34.3824, "lon": 50.5235, "population": 1.2},
    {"name": "Ahvaz", "lat": 31.1817, "lon": 48.4042, "population": 1.3},
    {"name": "Kermanshah", "lat": 34.2013, "lon": 47.0528, "population": 0.94},
    {"name": "Oromia", "lat": 37.5527, "lon": 45.0761, "population": 0.35}
]
```

```python
plt.figure(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=25, urcrnrlat=40,
            llcrnrlon=44, urcrnrlon=63, lat_ts=20, resolution='l')

m.drawcoastlines()
m.fillcontinents(color='orange', lake_color='aqua')
m.drawparallels(np.arange(20, 50, 5), labels=[1, 0, 0, 0])
m.drawmeridians(np.arange(40, 65, 5), labels=[0, 0, 0, 1])
m.drawmapboundary(fill_color='aqua')

lons = [city["lon"] for city in cities]
lats = [city["lat"] for city in cities]
sizes = [city["population"] * 100 for city in cities]

x, y = m(lons, lats)

plt.scatter(x, y, s=sizes, alpha=0.6, c='red', edgecolors='black')

for i in range(len(cities)):
    plt.text(x[i], y[i], cities[i]["name"], fontsize=10, ha='left', color='black')

plt.title("10 Most Populous Cities in Iran")
plt.show()
```
